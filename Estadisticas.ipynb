```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

Experimentos={}

for ex in ['HC','EE','DE']:
    mejores=[]
    for i in range(30):
        print '*',
        df = pd.read_csv('Resultados/'+ex+'_'+str(i)+'.txt',sep=',', header = None)
        mejores.append(df[0])
    data = pd.concat(mejores, axis=1)
    Experimentos[ex]=data
```

```python
def plot_prob(title,data):
    print title,"*************************************"
    result = pd.DataFrame()
    result['std']=data.std(axis=1)
    result['std'].plot();
    plt.title('Desviacion estandar')
    plt.grid(True)
    plt.show()

    result[title]=data.median(axis=1)
    result['median']=data.median(axis=1)
    result['max']=data.max(axis=1)
    result['min']=data.min(axis=1)

    result[['median','max','min']].plot();
    plt.title('Mediana')
    plt.grid(True)
    plt.show()

    result['mean']=data.mean(axis=1)
    result['mean+std']=result['mean']+result['std']
    result['mean-std']=result['mean']-result['std']

    result[['mean','mean+std','mean-std']].plot();


    plt.title('Promedio +- desviacion')
    plt.grid(True)
    plt.show()
    
    
    print data.median().median(), '+-', data.std().std()
    print data.mean().mean(), '+-', data.std().std()
    print result[['median','max','min','mean','std']]
    return result[title]
    
Meds=[]
for key in Experimentos:
    median=plot_prob(key,Experimentos[key])
    Meds.append(median)
```

```python
pltg=pd.concat(Meds, axis=1)
pltg.plot();
plt.show()    
```
